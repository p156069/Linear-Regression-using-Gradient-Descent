```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt


def gradient_descent_calculator(theta0, theta1, features, alpha):
    partial_theta0 = 0
    partial_theta1 = 0
    N = float(len(features))
    for i in range(0, len(features)):
        partial_theta0 += -(2/N) * (features[i, 1] - ((theta1 * features[i, 0]) + theta0))
        partial_theta1 += -(2/N) * features[i, 0] * (features[i, 1] - ((theta1 * features[i, 0]) + theta0))
    new_theta0 = theta0 - (alpha * partial_theta0)
    new_theta1 = theta1 - (alpha * partial_theta1)
    return [new_theta0, new_theta1]

def gradient_descent(features, theta0, theta1, alpha, iteration):
    new_theta1 = theta0
    new_theta1 = theta1
    for i in range(iteration):
        new_theta0, new_theta1 = gradient_descent_calculator(new_theta1, new_theta1, array(features), alpha)
    return [new_theta0, new_theta1]

def calculate_cost(theta0, theta1, features):
    cost = 0
    for i in range(0, len(features)):
        cost += (features[i, 1] - (theta1 * features[i, 0] + theta0)) ** 2
    return cost / float(len(features))


#Main

features = np.loadtxt("dataset.txt",dtype=np.float64,delimiter="|")
X = features[::,0]
Y = features[::,1]
theta0 = 0.0 
theta1 = 0.0 
alpha = 0.00000001
iteration = 1000

print ("At b = ",theta0,"m = ",theta1," Cost = ",calculate_cost(theta0, theta1, features))
[new_theta0, new_theta1] = gradient_descent(features, theta0, theta1, alpha, iteration)
print ("After ", iteration," b = ", new_theta0, " m =", new_theta0, " cost = ", calculate_cost(new_theta0, new_theta1, features))
    
plt.figure(figsize = (20,5),dpi=100)
plt.subplot(121)
plt.scatter(X,Y)
plt.xlabel("Size of house (X - axis)")
plt.ylabel("Price (Y - axis)")
plt.plot([0, 4000], [new_theta0, 4070*new_theta1+new_theta0], '-')
plt.show()
    
```
